```python
# Prompt for user input
name = input("Enter your name: ")
print(f"Hello {name}")

father = input("Enter your father's name: ")
roll = input("Enter your roll number: ")

# Define the correct expected values for comparison
correct_father_name = "michel"
correct_roll_number = "234" # Input is a string, so compare with a string

# Check conditions
if father.lower() == correct_father_name and roll == correct_roll_number:
    print("Correct")
    print("You registration completed")
else:
    # Handle incorrect inputs
    print("Wrong")
    if father.lower() != correct_father_name:
        print("Incorrect father's name. Please try again.")
    if roll != correct_roll_number:
        print("Incorrect roll number. Please try again.")

```

## **Input Number**
The input from the user is treated as a string. Even if, in the example above, you can input a number, the Python interpreter will still treat it as a string.

You can convert the input into a number with the **float()** function:

```python
import math
x = input("Enter a number:")

#find the square root of the number:
y = math.sqrt(float(x))

print(f"The square root of {x} is {y}")
```

```python
x= input("enter you number ")
y = math.acos(float(x))
print(f"the valu of cos of {x} is {y}")
```
