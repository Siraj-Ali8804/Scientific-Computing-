# **Python Output / Print**
print() function to display text or output values:

**Double Quotes**
Text in Python must be inside quotes. You can use either **"** double quotes or **'** single quotes:

```python
print("My name is siraj")
print('I am learning scientific computing')
```

**Print Without a New Line**

By default, the print() function ends with a new line.
If you want to print multiple words on the same line, you can use the **end** parameter:

```python
print("My name is siraj", end="")
print( " I am learning scientific computing")
```

```python
print("I have started learning python", end=" ")
print("my aim to be computational scientist")
```

```python
print("I will learn python basics in a month", end=" ")
print(" then i will start learning data analaysis in next month")
```
