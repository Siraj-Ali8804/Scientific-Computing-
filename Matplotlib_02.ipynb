```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
xx, yy = np.meshgrid(x, y)
z = xx**2 + yy**2

# FIX: Unpack both fig and ax from subplots directly
fig, ax = plt.subplots(figsize=(9, 6))

# Now ax is the Axes object, which has the contour method
p = ax.contour(xx, yy, z, cmap="winter")

fig.colorbar(p)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
xx, yy = np.meshgrid(x, y)
z = xx**2 + yy**2

# FIX: Unpack both fig and ax from subplots directly
fig, ax = plt.subplots(figsize=(9, 6))

# Now ax is the Axes object, which has the contour method
p = ax.contourf(xx, yy, z, cmap="winter")

fig.colorbar(p)
plt.show()
```

##Heat Map

```python
delivery = pd.read_csv("/content/IPL_Ball_by_Ball_2008_2022.csv")
```

```python
delivery.head()
```

```python
tem_df=delivery[delivery["ballnumber"].isin([1,2,3,4,5,6]) & (delivery["batsman_run"]==6)]
```

```python
grid= tem_df.pivot_table(index="overs",columns="ballnumber",values= "batsman_run",aggfunc="count")
```

```python
plt.figure(figsize=(8,8))
plt.imshow(grid)
plt.colorbar()
```
