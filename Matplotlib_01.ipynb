```python
plt.plot(batsman["index"],batsman["V Kohli"],color = "#3498eb")
plt.plot(batsman["index"],batsman["RG Sharma"],color= "#40eb34")
plt.title("comparisiom kholi vs sharma")
plt.xlabel("year",color="g")
plt.ylabel("score",color= "r")
```

```python
plt.plot(batsman["index"],batsman["V Kohli"],color = "#3498eb", linestyle="solid",linewidth=2)
plt.plot(batsman["index"],batsman["RG Sharma"],color= "#40eb34",linestyle= "dotted")
plt.title("comparisiom kholi vs sharma")
plt.xlabel("year",color="g")
plt.ylabel("score",color= "r")
```

```python
plt.plot(batsman["index"],batsman["V Kohli"],color = "#3498eb", linestyle="solid",linewidth=2,marker="d")
plt.plot(batsman["index"],batsman["RG Sharma"],color= "#40eb34",linestyle= "dotted",marker="s")
plt.title("comparisiom kholi vs sharma")
plt.xlabel("year",color="g")
plt.ylabel("score",color= "r")
```

```python
plt.plot(batsman["index"],batsman["V Kohli"],color = "#3498eb", linestyle="solid",linewidth=2,marker="d")
plt.plot(batsman["index"],batsman["RG Sharma"],color= "#40eb34",linestyle= "dotted",marker="s",markersize=10)
plt.title("comparisiom kholi vs sharma")
plt.xlabel("year",color="g")
plt.ylabel("score",color= "r")
```

```python
#legend
plt.plot(batsman["index"],batsman["V Kohli"],color = "#3498eb", linestyle="solid",linewidth=2,marker="d",label="rohit")
plt.plot(batsman["index"],batsman["RG Sharma"],color= "#40eb34",linestyle= "dotted",marker="s",markersize=10,label="virat")
plt.title("comparisiom kholi vs sharma")
plt.xlabel("year",color="g")
plt.ylabel("score",color= "r")
plt.legend()
```

```python
#plotting a simple funtion
price = [48000,540000,570000,470000,45000000]
year =[2012,2014,2017,2018,2019]
plt.plot(year,price)
plt.grid(True)
plt.ylim(0,3400000)
plt.xlim()
```

```python
#legend
plt.plot(batsman["index"],batsman["V Kohli"],color = "#3498eb", linestyle="solid",linewidth=2,marker="d",label="rohit")
plt.plot(batsman["index"],batsman["RG Sharma"],color= "#40eb34",linestyle= "dotted",marker="s",markersize=10,label="virat")
plt.title("comparisiom kholi vs sharma")
plt.xlabel("year",color="g")
plt.ylabel("score",color= "r")
plt.legend()
plt.grid(True)
plt.show()
```

```python
df = pd.read_csv("/content/batsman_season_record.csv")
df
```

```python
plt.bar(df["batsman"],df["2015"])
plt.bar(df["batsman"],df["2016"])
plt.bar(df["batsman"],df["2017"])
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df["2015"],width=0.2,color="r")
plt.bar(np.arange(df.shape[0]),df["2016"],width=0.2,color="b")
plt.bar(np.arange(df.shape[0])+0.2,df["2017"],width=0.2,color="g")
plt,plt.xticks(np.arange(df.shape[0]),df["batsman"])
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df["2015"],width=0.2,color="r")
plt.bar(np.arange(df.shape[0]),df["2016"],width=0.2,color="b")
plt.bar(np.arange(df.shape[0])+0.2,df["2017"],width=0.2,color="g")
plt,plt.xticks(np.arange(df.shape[0]),df["batsman"])
plt.show()
```
