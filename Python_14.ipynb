# Python Loops
oops are control flow statements used to repeatedly execute a block of code to automate repetitive tasks.

Python has two primitive loop commands:

**while** loops and
**for** loops

**while Loop**

With the while loop we can execute a set of statements as long as a condition is true.

```python
i = 1
while i < 6:
  print(i)
  i += 1
```

```python
i = 4
while i< 10:
  print(i)
  i=i+7
```

```python
a = 2
while a< 20:
  print(a)
  a+=1
```

```python
s=3
while s<20:
  print(s)
  s+=3
```

```python
m = 4
while m<20:
  print(m)
  m+=4
```

## The break Statement
With the **break** statement we can stop the loop even if the while condition is true:

```python
i = 1
while i < 6:
  print(i)
  if i == 5:
      break
  i += 1
```

```python
i = 1
while i<15:
  print(i)
  if i==10:
    break
  i+=2
```

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```

```python
a= 2
while a < 10:
  a+= 2
  if a== 4:
    continue
  print(a)
```

```python
e= 10
while e<40:
  e+=10
  if e == 30:
    continue
  print(e)
```

With the **else** statement we can run a block of code once when the condition no longer is true:

```python
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```
