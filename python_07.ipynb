# **Python Lists**
Lists are used to store multiple items in a single variable.

Lists are created using square brackets [ ].

List items are **ordered, changeable, and allow duplicate values**.

List items can be of any data type:

```python
my_list = [1,2,"a,s,d,F,F","abacus"]
print(my_list)
print(len(my_list))
```

```python
list_1 =["apple","banana","orange"]
print(len(list_1))
print(list_1)
```

**The list() Constructor**
It is also possible to use the **list()** constructor when creating a new list.

```python
list_2 = list(("apple","banana","orange"))
print(list_2)
print(len(list_2))
```

**Access Items**
List items are indexed and you can access them by referring to the index number:

```python
list_3=["my","name","is" ,24]
print(list_3[1:])
print(list_3[:1])
print(list_3[1:3])
print(list_3[-1:])
print(list_3[:-1])
```

**Check if Item Exists**
To determine if a specified item is present in a list use the **in** keyword:

```python
list_4= ["onioin", "salt","oil","garlic","ginger","tomamto","pepe","meat"]
if "water" in list_4:
  print("yes")
else:
    print("no")
```

**Insert Items**
To insert a new list item, without replacing any of the existing values, we can use the insert() method.

The **insert()** method inserts an item at the specified index:

**Append Items**
To add an item to the end of the list, use the append() method:

```python
list_10=['water', 'mirch', 'salt', 'oil', 'aplha', 'orange', 'banana', 'tomamto', 'pepe', 'meat']
list_10.append("cook")
print(list_10)
```

**Extend List**
To append elements from another list to the current list, use the **extend()** method.

```python
#Add the elements of tropical to thislist:
thislist = ["apple", "banana", "cherry"]
tropical = ["mango", "pineapple", "papaya"]
thislist.extend(tropical)
print(thislist)
```

```python
list_9=['water', 'mirch', 'salt', 'oil', 'aplha', 'orange', 'banana']
list_11=['apple', 'banana', 'cherry', 'mango', 'pineapple', 'papaya']
list_9.extend(list_11)
print(list_9)
```

```python
list_12 =['water', 'mirch', 'salt', 'oil', 'aplha', 'orange', 'banana']
list_13 =['apple', 'banana', 'cherry', 'mango', 'pineapple', 'papaya']
list_12.extend(list_13)
print(list_12)
```

**Remove Specified Item**

**Remove Specified Index**
The **pop()** method removes the specified index.

**Customize Sort Function**
You can also customize your own function by using the keyword argument **key = function**.

The function will return a number that will be used to sort the list (the lowest number first):

```python
#Sort the list based on how close the number is to 50:

def myfunc(n):
  return abs(n - 50)

thislist = [100, 50, 65, 82, 23]
thislist.sort(key = myfunc)
print(thislist)
```

**Case Insensitive Sort**
By default the **sort()** method is case sensitive, resulting in all capital letters being sorted before lower case letters:

```python
thislist = ["banana", "Orange", "Kiwi", "Cherry"]
thislist.sort()
print(thislist)
```

**Reverse Order**
What if you want to reverse the order of a list, regardless of the alphabet?

The **reverse()** method reverses the current sorting order of the elements.

```python
thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.reverse()
print(thislist)
```

```python
list_17 = ['salt', 'aplha', 'orange', 'banana', 'pineapple', 'papaya']
list_17.reverse()
print(list_17)
```
