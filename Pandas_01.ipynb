series from lists

```python
import pandas as pd
country= ["india","Pakistan","USA","Nepal"]
pd.Series(country)
```

```python
runs= [12,35,5,55,6]
pd.Series(runs)
```

```python
#costum index
marks =[34,56,67,33]
subjects=["maths",'URDU',"hindi","science"]
pd.Series(marks,index=subjects)
```

```python
#setting
pd.Series(marks,index=subjects,name= 'MY MARKS')
```

```python
marks={'maths':33,"englis":64,"science":44}
marks_series= pd.Series(marks,name="my marks")
marks_series
```
